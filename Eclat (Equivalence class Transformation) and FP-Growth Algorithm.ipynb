```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
dataset=[['Bread','Butter','Olive Oil','Veggies','Tomatoes','Broccoli'],
         ['Grapes','Butter','Olive Oil','Veggies','Tomatoes','Broccoli'],
         ['Bread'],
         ['Bread','Olives','Fish','Veggies','Broccoli'],
         ['Fish','Butter','Butter','Veggies','Soy Sauce','Tomatoes']
         ]
```

```python
dataset
```

```python
freq_itemsets=pd.DataFrame(dataset)
freq_itemsets
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)

df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python

frequent_itemsets=apriori(df,min_support=0.6,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='support',support_only=True,min_threshold=0.1)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support']]
rules
```
